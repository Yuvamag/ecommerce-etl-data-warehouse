```python
# ============================================
# FINAL PROFESSIONAL EDA PROFILING SCRIPT
# ============================================

import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.backends.backend_pdf import PdfPages

# -----------------------------
# SETTINGS
# -----------------------------
file_path = "dataset.csv"
output_folder = "eda_profiling_final"
os.makedirs(output_folder, exist_ok=True)

# -----------------------------
# LOAD DATA
# -----------------------------
df = pd.read_csv(file_path, encoding='ISO-8859-1')

# Standardize types
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], errors='coerce')

df = df.astype({
    'Order ID': 'string',
    'Customer ID': 'string',
    'Product ID': 'string'
})

print("Dataset Shape:", df.shape)

# ============================================
# COLUMN PROFILING SUMMARY
# ============================================

summary_list = []

for col in df.columns:
    null_count = df[col].isnull().sum()
    null_percent = (null_count / len(df)) * 100
    distinct_count = df[col].nunique()
    duplicate_count = len(df) - distinct_count
    duplicate_percent = (duplicate_count / len(df)) * 100
    dtype = df[col].dtype
    
    min_val = max_val = mean_val = std_val = None
    
    if pd.api.types.is_numeric_dtype(df[col]):
        min_val = df[col].min()
        max_val = df[col].max()
        mean_val = df[col].mean()
        std_val = df[col].std()
    
    top_freq = df[col].value_counts().head(5).to_dict()
    
    summary_list.append([
        col, dtype, null_count, null_percent,
        distinct_count, duplicate_count, duplicate_percent,
        min_val, max_val, mean_val, std_val,
        str(top_freq)
    ])

column_summary = pd.DataFrame(summary_list, columns=[
    "Column", "Data_Type", "Null_Count", "Null_%", 
    "Distinct_Count", "Duplicate_Count", "Duplicate_%",
    "Min", "Max", "Mean", "Std_Dev",
    "Top_5_Frequencies"
])

column_summary.to_csv(f"{output_folder}/column_summary_final.csv", index=False)

# ============================================
# DUPLICATE ANALYSIS
# ============================================

full_row_duplicates = df.duplicated().sum()
composite_duplicates = df.duplicated(
    subset=['Order ID','Product ID']
).sum()

# ============================================
# KEY VALIDATION
# ============================================

key_tests = []

def check_uniqueness(columns, name):
    dup_count = df.duplicated(subset=columns).sum()
    is_unique = dup_count == 0
    key_tests.append([name, is_unique, dup_count])

check_uniqueness(['Row ID'], 'Row ID')
check_uniqueness(['Order ID'], 'Order ID')
check_uniqueness(['Product ID'], 'Product ID')
check_uniqueness(['Customer ID'], 'Customer ID')
check_uniqueness(['Order ID','Product ID'], 'OrderID + ProductID')
check_uniqueness(['Order ID','Product ID','Row ID'], 'OrderID + ProductID + RowID')

key_validation = pd.DataFrame(key_tests, columns=[
    "Key_Test", "Is_Unique", "Duplicate_Count"
])

key_validation.to_csv(f"{output_folder}/key_validation_final.csv", index=False)

# ============================================
# BUSINESS RULE CHECKS
# ============================================

business_rules = {
    "Negative Sales": (df['Sales'] < 0).sum(),
    "Negative Profit": (df['Profit'] < 0).sum(),
    "Discount > 1": (df['Discount'] > 1).sum(),
    "Discount < 0": (df['Discount'] < 0).sum(),
    "Ship Date before Order Date": (df['Ship Date'] < df['Order Date']).sum()
}

business_df = pd.DataFrame(list(business_rules.items()),
                           columns=["Rule", "Violation_Count"])

business_df.to_csv(f"{output_folder}/business_rule_checks_final.csv", index=False)

# ============================================
# CORRELATION HEATMAP (Selective Metrics)
# ============================================

numeric_cols = ['Sales','Profit','Discount','Quantity']
corr_df = df[numeric_cols]

corr_matrix = corr_df.corr()

plt.figure(figsize=(6,5))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Selective Correlation Heatmap")
plt.tight_layout()
plt.savefig(f"{output_folder}/correlation_heatmap_final.png")
plt.close()

# ============================================
# HIGH DUPLICATE COLUMN ANALYSIS (>50%)
# ============================================

high_dup = column_summary[column_summary["Duplicate_%"] > 50]
high_dup.to_csv(f"{output_folder}/high_duplicate_columns_final.csv", index=False)

if not high_dup.empty:
    plt.figure(figsize=(10,5))
    sns.barplot(x="Column", y="Duplicate_%", data=high_dup)
    plt.xticks(rotation=45, ha="right")
    plt.title("Columns with >50% Duplicate Values")
    plt.tight_layout()
    plt.savefig(f"{output_folder}/high_duplicate_columns_chart_final.png")
    plt.close()

# ============================================
# SELECTIVE CARDINALITY GRAPH
# ============================================

selected_cols = ['Order ID','Product ID','Customer ID','Region','Category']
for col in selected_cols:
    distinct = df[col].nunique()
    duplicates = len(df) - distinct
    duplicate_entries = df[col].duplicated().sum()
    
    plt.figure(figsize=(5,4))
    plt.bar(["Unique", "Duplicates", "Duplicate Value Entries"], 
            [distinct, duplicates, duplicate_entries], color=['steelblue','green','pink'])
    plt.title(f"Cardinality: {col}")
    plt.tight_layout()
    plt.savefig(f"{output_folder}/cardinality_{col}_final.png")
    plt.close()

# ============================================
# PDF REPORT
# ============================================

pdf_path = f"{output_folder}/eda_profile_full_report_final.pdf"

with PdfPages(pdf_path) as pdf:
    
    # Page 1 – Overview
    plt.figure(figsize=(10,6))
    plt.axis('off')
    plt.text(0.05,0.8,"DATASET OVERVIEW", fontsize=16)
    plt.text(0.05,0.7,f"Shape: {df.shape}")
    plt.text(0.05,0.6,f"Full Row Duplicates: {full_row_duplicates}")
    plt.text(0.05,0.5,f"Composite Key Duplicates: {composite_duplicates}")
    plt.text(0.05,0.4,"Grain: One row per Order Line (Row ID)")
    pdf.savefig()
    plt.close()
    
    # Page 2 – Correlation Heatmap
    img = plt.imread(f"{output_folder}/correlation_heatmap_final.png")
    plt.figure(figsize=(8,6))
    plt.imshow(img)
    plt.axis('off')
    pdf.savefig()
    plt.close()
    
    # Page 3 – High Duplicate Chart
    if not high_dup.empty:
        img = plt.imread(f"{output_folder}/high_duplicate_columns_chart_final.png")
        plt.figure(figsize=(8,6))
        plt.imshow(img)
        plt.axis('off')
        pdf.savefig()
        plt.close()
    
    # Page 4+ – Cardinality Graphs for Selected Columns
    for col in selected_cols:
        img = plt.imread(f"{output_folder}/cardinality_{col}_final.png")
        plt.figure(figsize=(6,5))
        plt.imshow(img)
        plt.axis('off')
        pdf.savefig()
        plt.close()

print("✅ FINAL MASTER PROFILING COMPLETE")
print("All outputs saved in:", output_folder)
```

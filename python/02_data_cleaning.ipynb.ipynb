```python
import pandas as pd

df = pd.read_csv('dataset.csv', encoding='ISO-8859-1')
```

```python
# Remove unwanted index column if present
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
```

```python
#trim whitespace
text_cols = df.select_dtypes(include=['object', 'string']).columns

for col in text_cols:
    df[col] = df[col].str.strip()
```

```python
#Standardize casing
title_case_cols = [
    'Customer Name', 'City', 'State', 'Country',
    'Region', 'Category', 'Sub-Category', 'Product Name'
]

for col in title_case_cols:
    df[col] = df[col].str.title()

df['Ship Mode'] = df['Ship Mode'].str.title()
df['Segment'] = df['Segment'].str.title()
```

```python
#enforcing data types
df = df.astype({
    'Row ID': 'int64',
    'Order ID': 'string',
    'Customer ID': 'string',
    'Product ID': 'string',
    'Postal Code': 'int64',
    'Quantity': 'int64',
    'Sales': 'float64',
    'Discount': 'float64',
    'Profit': 'float64'
})

df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

```python
df.to_csv("cleaned_sales.csv", index=False)
```
